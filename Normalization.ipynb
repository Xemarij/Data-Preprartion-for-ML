# **Min-Max Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
!cd heart-disease-statlog
```

```python
Heart_Disease_Data = pd.read_csv('/content/heart-disease-statlog/Heart_disease_statlog.csv')
```

```python
Heart_Disease_Data.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(Heart_Disease_Data['chol'])
plt.show()
```

```python
Heart_Disease_Data['chol'].describe()
```

```python
Heart_Disease_Data_MM = MinMaxScaler().fit_transform(Heart_Disease_Data[['chol']])
```

```python
Heart_Disease_Data_MM
```

```python
Heart_Disease_Data_MM = pd.DataFrame(Heart_Disease_Data_MM, columns = ["chol"])
```

```python
Heart_Disease_Data_MM
```

```python
Heart_Disease_Data_MM.describe()
```

# Here you can see that the min value is 0 and the max value is 1 which is the objective of nrmalization.Moreover the basic shape of the histogram before and after normalization does not change.

```python
import matplotlib.pyplot as plt
plt.hist(Heart_Disease_Data_MM)
plt.show()
```

# **Z-score Normalization (Standardization)**

```python
from sklearn.preprocessing import StandardScaler
```

```python
Heart_Disease_Data_SS = StandardScaler().fit_transform(Heart_Disease_Data[['chol']])
```

```python
Heart_Disease_Data_SS
```

```python
df = pd.DataFrame(Heart_Disease_Data_SS, columns = ["chol"])
```

```python
df
```

```python
df.describe()
```

```python
hist = plt.hist(df)
plt.show()
```

# It can be observed that the standardization or the z-score brings the mean to zero nad the standard deviation is set to one which is the objective of this kind of Normalization.
